```python
import pandas as pd
from sqlalchemy import create_engine
import chardet
```

```python
engine = create_engine(
    "postgresql+psycopg2://postgres:akasmik@localhost:5432/pizzaProject"
)
```

```python
def read_csv_safe(path):
    with open(path,"rb") as f:
        enc = chardet.detect(f.read(100000))["encoding"]
    print(f"Reading {path} with encoding={enc}")
    return pd.read_csv(
        path,
        engine = "python",
        encoding = enc
    )
```

```python
csv_table_flow = [
    ("pizza_types.csv","pizza_types"),
    ("pizzas.csv","pizzas"),
    ("orders.csv","orders"),
    ("order_details.csv","order_details"),
]

for csv, table in csv_table_flow:
    df = read_csv_safe(csv)
    df.columns = df.columns.str.strip()

    if table == "orders":
        df['date'] = pd.to_datetime(df['date']).dt.date
        df['time'] = pd.to_datetime(df['time']).dt.time
    df.to_sql(table,engine,if_exists='append',index=False,method='multi')
print("All data loaded successfully")
```
